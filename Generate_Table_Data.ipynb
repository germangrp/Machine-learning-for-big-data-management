## Importar librerias

```python
import pandas as pd
import numpy as np
```

## Carga y comprobaci√≥n de CSV

```python
# 1. Load the Dataset
df = pd.read_csv('cars.csv')

# 2. Inspect the Dataset
print(df.head())
print(df.dtypes)
```

## Llenar datos faltantes de"hp" con promedio

```python
# 3. Handle Missing Values
df['hp'].fillna(df['hp'].mean(), inplace=True)
```

## Eliminar renglones con "hp" faltante

```python
df.dropna(subset=['hp'], inplace=True)
```

## Nuevo Dataframe solo con cyl=6

```python
# 4. Filter Data
df_filtered = df[df['cyl'] == 6]
print(df_filtered)
```

## Categorizar "hp" en bajo, medio y bajo

```python
# 5. Transform Columns
def categorize_hp(hp):
    if hp > 150:
        return 'High'
    elif hp > 80:
        return 'Medium'
    else:
        return 'Low'
```

## Crear columna nueva

```python
df_filtered['HP_Category'] = df_filtered['hp'].apply(categorize_hp)
```

## Guardar tabla nueva

```python
# 6. Save the Cleaned Dataset
df_filtered.to_csv('.\dataset\cleaned_cars_dataset.csv', index=False)
```
